```python
x = int(input("Input number: ")) 

epsilon = 0.01
ans=0.0

while abs(ans**2-x) >= epsilon:     
    print (f"Current square root is {ans:.2f}")
    ans +=0.01
    if ans**2 + epsilon > x :
        break
    
if abs(ans**2-x) > epsilon:
    print("Failed on square root of",x)
else:
    print (x, "'s square root is", round(ans,1))
```
